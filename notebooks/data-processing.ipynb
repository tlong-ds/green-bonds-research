## Econometric Data Cleaning
Applying data cleaning techniques based on the Econometric Data Cleaning skill and the Evaluation Report fixes.

**Pipeline order (corrected):**
1. Load data & filter non-ASEAN firms / Year 2025
2. Handle missing values via forward fill (grouped by `ric`), then drop firms with <3 years of `total_assets`
3. Data type conversion (ensure numeric columns are numeric early)
4. Currency conversion to USD using `yfinance` (excludes ratios: ROA, ROE, emissions_intensity)
5. Winsorization at 1st/99th percentiles (AFTER currency conversion, in common currency)
6. Normalize ROA/ROE from percentage to decimal (divide by 100)
7. Log transform `total_assets`
8. Encode `environmental_investment` as binary (Y=1, N=0, NaN=NaN)
9. Data type verification & duplicate resolution


```python
import pandas as pd
import yfinance as yf
from scipy.stats.mstats import winsorize
import numpy as np

# Load data

# ============================================================
# 0. Filter non-ASEAN firms and Year 2025
# (Issue #7, #8, #18: "Other" country includes Korean firms
#  in non-ASEAN exchanges; Year 2025 has almost all NaN financials)
# ============================================================
final_panel = final_panel[final_panel['country'] != 'Other'].copy()
print(f"Shape after removing non-ASEAN ('Other') firms: {final_panel.shape}")

final_panel['Year'] = pd.to_numeric(final_panel['Year'], errors='coerce')
final_panel = final_panel[final_panel['Year'] != 2025].copy()
print(f"Shape after removing Year 2025: {final_panel.shape}")

# ============================================================
# 1. Handle Missing Values
# Forward fill slowly-changing variables, grouped by 'ric'.
# NOTE: environmental_investment is NOT forward-filled (Issue #19)
#   — it is a Y/N indicator, ffill would artificially inflate 'Y'.
# ============================================================
# Financial variables: unlimited forward fill (slowly-changing)
financial_ffill_vars = [
    'total_assets', 'return_on_equity_total', 'return_on_assets',
    'internal_carbon_pricing', 'internal_carbon_price_per_tonne'
]
for col in financial_ffill_vars:
    if col in final_panel.columns:
        final_panel[col] = final_panel.groupby('ric')[col].transform(lambda x: x.ffill())

# Environmental/ESG variables: limit ffill to 1 year (Issue L from eval v3)
# These change year-over-year; unlimited ffill creates artificial stability
# that biases DID analysis on green bond timing effects
esg_ffill_vars = [
    'emissions_intensity', 'esg_score', 'estimated_total_carbon_footprint'
]
for col in esg_ffill_vars:
    if col in final_panel.columns:
        final_panel[col] = final_panel.groupby('ric')[col].transform(lambda x: x.ffill(limit=1))

# Reduce data loss (Issue #15): keep firms with at least 3 years of
# non-null total_assets, then drop individual rows still missing it.
ta_notna_count = final_panel.groupby('ric')['total_assets'].transform(
    lambda x: x.notna().sum()
)
final_panel = final_panel[ta_notna_count >= 3].copy()
final_panel = final_panel.dropna(subset=['total_assets']).copy()
print(f"Shape after ffill and filtering firms with <3 years of total_assets: {final_panel.shape}")

# ============================================================
# 2. Early Data Type Conversion (Issue #9)
# Convert numeric columns from object to float BEFORE any math.
# ============================================================
numeric_cols = [
    'ask_price', 'bid_price', 'capital_expenditures', 'cash',
    'current_assets_total', 'current_liabilities_total',
    'earnings_bef_interest_tax', 'interest_expense_total',
    'long_term_debt', 'market_capitalization', 'market_value',
    'net_cash_flow_operating_actv', 'net_sales_or_revenues',
    'operating_income', 'total_assets', 'total_capital',
    'total_debt', 'total_liabilities',
    'return_on_assets', 'return_on_equity_total',
    'emissions_intensity', 'esg_score', 'estimated_total_carbon_footprint',
    'internal_carbon_pricing', 'internal_carbon_price_per_tonne',
    'employees', 'tri'
]
for col in numeric_cols:
    if col in final_panel.columns:
        final_panel[col] = pd.to_numeric(final_panel[col], errors='coerce')

# ============================================================
# 3. Currency Conversion to USD (BEFORE winsorization)
# (Issue #16: winsorize AFTER conversion to common currency)
# (Issue #17: EXCLUDE ratios from FX conversion — ROA, ROE,
#  emissions_intensity are percentages/ratios, not currency values)
# ============================================================
currency_map = {
    'Vietnam': 'VND=X',
    'Thailand': 'THB=X',
    'Malaysia': 'MYR=X',
    'Singapore': 'SGD=X',
    'Indonesia': 'IDR=X',
    'Philippines': 'PHP=X'
}

fx_rates = {}
for country, ticker in currency_map.items():
    tk = yf.Ticker(ticker)
    hist = tk.history(period="max")
    if not hist.empty:
        hist['Year'] = hist.index.year
        yearly_avg = hist.groupby('Year')['Close'].mean()
        fx_rates[country] = yearly_avg.to_dict()

# Only MONETARY columns — explicitly exclude ratios/percentages
financial_cols_for_fx = [
    'ask_price', 'bid_price', 'capital_expenditures', 'cash',
    'current_assets_total', 'current_liabilities_total',
    'earnings_bef_interest_tax', 'interest_expense_total',
    'long_term_debt', 'market_capitalization', 'market_value',
    'net_cash_flow_operating_actv', 'net_sales_or_revenues',
    'operating_income', 'total_assets', 'total_capital',
    'total_debt', 'total_liabilities'
]
# NOTE: return_on_assets, return_on_equity_total, emissions_intensity,
#        esg_score, internal_carbon_pricing are NOT converted.

def get_fx_rate(row):
    country = row['country']
    year = row['Year']
    if country in fx_rates and year in fx_rates[country]:
        return fx_rates[country][year]
    return 1.0  # fallback (should not happen after filtering 'Other')

final_panel['fx_rate'] = final_panel.apply(get_fx_rate, axis=1)

for col in financial_cols_for_fx:
    if col in final_panel.columns:
        final_panel[col] = final_panel[col] / final_panel['fx_rate']

final_panel.drop(columns=['fx_rate'], inplace=True)
print("Currency conversion to USD complete (ratios excluded).")

# ============================================================
# 4. Winsorization (AFTER currency conversion — Issue #16)
# Now all monetary values are in USD, so percentile trimming
# is statistically meaningful across countries.
# Also winsorize emissions_intensity (Issue #25).
# ============================================================
winsorize_vars = [
    'ask_price', 'bid_price', 'total_assets', 'net_sales_or_revenues',
    'cash', 'capital_expenditures', 'long_term_debt', 'total_debt',
    'market_capitalization', 'market_value', 'operating_income',
    'earnings_bef_interest_tax', 'current_assets_total',
    'current_liabilities_total', 'total_liabilities', 'total_capital',
    'emissions_intensity', 'estimated_total_carbon_footprint'
]
# Also winsorize ratios separately
ratio_winsorize_vars = ['return_on_assets', 'return_on_equity_total']

for col in winsorize_vars + ratio_winsorize_vars:
    if col in final_panel.columns:
        mask = final_panel[col].notna()
        if mask.sum() > 0:
            final_panel.loc[mask, col] = winsorize(final_panel.loc[mask, col], limits=[0.01, 0.01])

print("Winsorization complete (post-FX conversion).")

# ============================================================
# 5. Normalize ROA and ROE from percentage to decimal
# (User feedback: these are stored as percentages, divide by 100)
# ============================================================
for col in ['return_on_assets', 'return_on_equity_total']:
    if col in final_panel.columns:
        final_panel[col] = final_panel[col] / 100.0

# ============================================================
# 6. Data Transformation & Normalization
# Log transform total_assets (standard in corporate finance)
# ============================================================
final_panel['ln_total_assets'] = np.log(final_panel['total_assets'].replace(0, np.nan))

# ============================================================
# 7. Encode environmental_investment properly (Issue #11)
# Map Y→1, N→0, keep NaN as NaN (not falsely coded as 0)
# ============================================================
final_panel = pd.get_dummies(final_panel, columns=["environmental_investment"], drop_first=True)
final_panel.columns = final_panel.columns.str.strip()

final_panel = final_panel.rename(columns={
    "environmental_investment_Y": "environmental_investment"
    })

print(f"environmental_investment encoded: "
        f"1={int((final_panel['environmental_investment']==1).sum())}, "
        f"0={int((final_panel['environmental_investment']==0).sum())}, "
        f"NaN={int(final_panel['environmental_investment'].isna().sum())}")

# ============================================================
# 8. Data Type Verification
# ============================================================
final_panel['isin'] = final_panel['isin'].astype(str)
final_panel['ric'] = final_panel['ric'].astype(str)
final_panel['country'] = final_panel['country'].astype(str)
final_panel['company'] = final_panel['company'].astype(str)
final_panel['Year'] = final_panel['Year'].astype(int)

# Treat GIC as categorical, not numeric (Issue #14)
if 'gic' in final_panel.columns:
    final_panel['gic'] = final_panel['gic'].astype(str).astype('category')

# ============================================================
# 9. Duplicates Resolution
# ============================================================
final_panel = final_panel.sort_values(by=final_panel.columns.tolist()).drop_duplicates(subset=['ric', 'Year'], keep='last')

print(f"\nData Cleaning Complete. Final Shape: {final_panel.shape}")
print(f"Countries: {final_panel['country'].unique()}")
print(f"Year range: {final_panel['Year'].min()} - {final_panel['Year'].max()}")
print(f"Unique firms (RICs): {final_panel['ric'].nunique()}")

# Save the cleaned dataset for the next step
final_panel.to_csv('../processed_data/cleaned_panel_data.csv', index=False)
```

## Variable Engineering
Calculations of intermediate financial and environmental variables for regression models.



```python
import pandas as pd
import numpy as np

# ============================================================
# 1. Calculate Control Variables
# ============================================================
# Firm_Size: proper copy, not a reference (Issue #26)
final_panel['Firm_Size'] = final_panel['ln_total_assets'].copy()

# Leverage: clip to [0, 1.5] — values > 1 can occur for distressed firms
# where debt exceeds assets, which is legitimate in ASEAN restructuring cases.
# Only extreme values (> 1.5) are treated as data errors (Issue #24, relaxed in eval v3)
final_panel['Leverage'] = (final_panel['total_debt'] / final_panel['total_assets']).clip(lower=0, upper=1.5)

# Capital_Intensity: cap at [0, 1] — CapEx/Assets > 1 is almost certainly
# a data error (Issue #27)
final_panel['Capital_Intensity'] = (final_panel['capital_expenditures'].abs() / final_panel['total_assets']).clip(lower=0, upper=1)

print("Control variables calculated:")
print(final_panel[['Firm_Size', 'Leverage', 'Capital_Intensity']].describe())

# ============================================================
# 2. Green Bond & ESG Processing
# ============================================================
# Log-transform emissions_intensity (extreme skewness = 33.5, Issue #25)
# Already winsorized in cleaning step; log calms remaining skewness
if 'emissions_intensity' in final_panel.columns:
    final_panel['ln_emissions_intensity'] = np.log(final_panel['emissions_intensity'].replace(0, np.nan))

# Log-transform estimated_total_carbon_footprint if present
if 'estimated_total_carbon_footprint' in final_panel.columns:
    final_panel['ln_carbon_footprint'] = np.log(
        final_panel['estimated_total_carbon_footprint'].replace(0, np.nan)
    )

# ============================================================
# 3. GIC as categorical (Issue #14)
# ============================================================
if 'gic' in final_panel.columns:
    final_panel['gic'] = final_panel['gic'].astype(str).astype('category')
    print(f"GIC categories: {final_panel['gic'].nunique()} unique industry codes")

# ============================================================
# 4. Country and Year Dummies (Fixed Effects Preparation)
# Note: 'Other' country already filtered in cleaning. Year 2025 removed.
# environmental_investment is already 1/0/NaN — no get_dummies needed for it.
# ============================================================
final_panel = pd.get_dummies(final_panel, columns=['country', 'Year'], drop_first=True, dtype=int)

```

```python
# Ensure Tobin's Q is available if not already present
if 'Tobin_Q' not in final_panel.columns and 'market_value' in final_panel.columns and 'total_liabilities' in final_panel.columns and 'total_assets' in final_panel.columns:
    final_panel['Tobin_Q'] = (final_panel['market_value'] + final_panel['total_liabilities']) / final_panel['total_assets']

# Engineer Ratios to avoid absolute scale multicollinearity and tautologies
ratio_cols = ['Cash_Ratio', 'CapEx_Ratio', 'Current_Ratio', 'Asset_Turnover', 'Operating_Margin']
final_panel['Cash_Ratio'] = (final_panel['cash'] / final_panel['total_assets']).replace([np.inf, -np.inf], np.nan)
final_panel['CapEx_Ratio'] = (final_panel['capital_expenditures'] / final_panel['total_assets']).replace([np.inf, -np.inf], np.nan)
final_panel['Current_Ratio'] = (final_panel['current_assets_total'] / final_panel['current_liabilities_total']).replace([np.inf, -np.inf], np.nan)
final_panel['Asset_Turnover'] = (final_panel['net_sales_or_revenues'] / final_panel['total_assets']).replace([np.inf, -np.inf], np.nan)
final_panel['Operating_Margin'] = (final_panel['operating_income'] / final_panel['net_sales_or_revenues']).replace([np.inf, -np.inf], np.nan)

# Apply robust winsorization (1st and 99th percentiles) on engineered ratios
for col in ratio_cols:
    lower = final_panel[col].quantile(0.01)
    upper = final_panel[col].quantile(0.99)
    final_panel[col] = final_panel[col].clip(lower=lower, upper=upper)


# Reconstruct Year for proper panel lagging of our new ratios
year_cols = [c for c in final_panel.columns if c.startswith('Year_')]
final_panel['Year'] = final_panel[year_cols].idxmax(axis=1).str.replace('Year_', '').astype(int)
final_panel = final_panel.sort_values(['company', 'Year'])

# Fix Panel Data Lags
# 1. Use GroupBy + shift(1) is dangerous if years are missing. We will create a Last_Year column and merge.
final_panel_lag = final_panel.copy()
final_panel_lag['Year'] = final_panel_lag['Year'] + 1 # Shift year forward to match with next year's data

financial_cands_base = [
    "Firm_Size", "Leverage", "Capital_Intensity",
    "Cash_Ratio", "CapEx_Ratio", "Current_Ratio", "Asset_Turnover", "Operating_Margin",
    "return_on_assets", "Tobin_Q"
]

environmental_cands_base = [
    "esg_score", "ln_emissions_intensity", 
    "environmental_investment", "ln_carbon_footprint"
]

cols_to_lag = ['company', 'Year'] + [c for c in financial_cands_base + environmental_cands_base if c in final_panel_lag.columns]
final_panel_lag = final_panel_lag[cols_to_lag]

# Rename columns to L1_
rename_dict = {c: f'L1_{c}' for c in financial_cands_base + environmental_cands_base if c in final_panel_lag.columns}
final_panel_lag = final_panel_lag.rename(columns=rename_dict)

# Drop existing L1 columns from final_panel to avoid _x and _y suffixing during merge
overlap_cols = [c for c in final_panel_lag.columns if c in final_panel.columns and c not in ['company', 'Year']]
final_panel = final_panel.drop(columns=overlap_cols, errors='ignore')

# Merge back on company and Year
final_panel = pd.merge(final_panel, final_panel_lag, on=['company', 'Year'], how='left')

final_panel.head(3)

final_panel.to_csv('../processed_data/final_engineered_panel_data.csv', index=False)
```
