```python
import pandas as pd
import numpy as np
```

# Merge Panel Data

```python
# 1. Load the data
df1 = pd.read_csv("data/panel_data.csv")
df2 = pd.read_csv("data/esg_panel_data.csv")

# 2. DROP THE EXTRA HEADER ROWS
df1 = df1[df1["Year"] != "Year"]
df2 = df2[df2["Year"] != "Year"]

# 3. NOW it is safe to convert them to integers
df1["Year"] = df1["Year"].astype(int)

# (df2's columns were renamed in your cell 12, so make sure to do that first)
df2.columns = ['company', 'isin', 'country', 'Year', 'emissions_intensity',
       'environmental_investment', 'internal_carbon_pricing']
df2["Year"] = df2["Year"].astype(int)
```

```python
df1.head()
```

```python
df2.head()
```

```python
# load market data
vn = pd.read_csv("data/vn-market.csv")
tl = pd.read_csv("data/tl-market.csv")
ml = pd.read_csv("data/ml-market.csv")
sing = pd.read_csv("data/sing-market.csv")
pp = pd.read_csv("data/pp-market.csv")
other = pd.read_csv("data/other-market.csv")
```

```python
market_data = pd.concat([vn, tl, ml, sing, pp], ignore_index=True)
market_data.drop("currency", axis=1, inplace=True)

other = other.reindex(columns=["name", "ric", "org_permid", "isin"])
market_data = pd.concat([market_data, other], ignore_index=True)
```

```python
tmp1 = pd.merge(df2, market_data, on="isin", how="inner")
tmp1.drop(["name", "org_permid", "company", "country"], axis=1, inplace=True)
```

```python
df1.columns = ['company', 'ric', 'country', 'Year', 'ask_price', 'bid_price',
       'capital_expenditures', 'cash', 'current_assets_total',
       'current_liabilities_total', 'earnings_bef_interest_tax', 'employees',
       'interest_expense_total', 'long_term_debt', 'market_capitalization',
       'market_value', 'net_cash_flow_operating_actv', 'net_sales_or_revenues',
       'operating_income', 'return_on_assets', 'return_on_equity_total',
       'total_assets', 'total_capital', 'total_debt', 'total_liabilities',
       'tri']
```

```python
df1["Year"] = df1["Year"].astype(int)
tmp1["Year"] = tmp1["Year"].astype(int)

merge_data = pd.merge(df1, tmp1, on=["ric", "Year"], how="left")
merge_data
```

```python
# Reset the index so it starts from 0 cleanly
merge_data = merge_data.reset_index(drop=True)

# Drop redundant duplicate rows if the market data caused any
merge_data = merge_data.drop_duplicates(subset=["ric", "Year"])

# Verify the final count 
print("Final Cleaned Panel Rows:", len(merge_data))
merge_data.head()
```

```python
gb = pd.read_csv("data/greenbonds.csv")
```

```python
gb.head()
```

```python
gb["Year"] = pd.to_datetime(gb["Dates: Issue Date"]).dt.year
# Keep only necessary columns
gb = gb[["Issuer/Borrower PermID", "Year", "Proceeds Amount This Market"]].rename(columns={"Issuer/Borrower PermID": "org_permid"})
# Standardize org_permid to string to match market_data
gb["org_permid"] = gb["org_permid"].apply(lambda x: str(int(x)) if pd.notna(x) else np.nan)
gb.head()
```

```python
market_data["org_permid_str"] = pd.to_numeric(market_data["org_permid"], errors='coerce')
market_data["org_permid_str"] = market_data["org_permid_str"].apply(lambda x: str(int(x)) if pd.notna(x) else np.nan)
# Map PermID to RIC
gb_mapped = pd.merge(gb, market_data[["org_permid_str", "ric"]].dropna(), left_on="org_permid", right_on="org_permid_str", how="inner")
gb_mapped.head()
```

```python
gb_agg = gb_mapped.groupby(["ric", "Year"], as_index=False).agg(
    green_bond_proceeds=("Proceeds Amount This Market", "sum")
)
gb_agg["green_bond_issue"] = 1
gb_agg.head()
```

```python
final_panel = pd.merge(merge_data, gb_agg, on=["ric", "Year"], how="left")
# Fill missing values from the merge with 0 (since no issuance = 0)
final_panel["green_bond_issue"] = final_panel["green_bond_issue"].fillna(0)
final_panel["green_bond_proceeds"] = final_panel["green_bond_proceeds"].fillna(0)
# Create the cumulative Green Bond dummy (1 for the year of issue and all years after)
final_panel = final_panel.sort_values(by=["ric", "Year"])
final_panel["green_bond_active"] = final_panel.groupby("ric")["green_bond_issue"].cumsum().clip(upper=1)
print(final_panel.shape)
final_panel.head()
```

```python
final_panel = final_panel.sort_index()
```

```python
final_panel
```

```python
# 1. Load the Series Data
series_df = pd.read_csv("data/series_data.csv")

# 2. We only need the 'ticker' (which is the ric) and 'gic' columns
# (You already have the 'country' column in your final_panel)
series_df = series_df[['ticker', 'gic']]

# 3. Rename 'ticker' to 'ric' so it perfectly matches the final_panel column name
series_df = series_df.rename(columns={'ticker': 'ric'})

# 4. Merge the GIC codes back into the master panel!
final_panel = pd.merge(final_panel, series_df, on="ric", how="left")

# Print to check the output
print(final_panel.shape)

# (Optional: If you appended multiple times, the series data might have duplicates too)
# Run this just to be perfectly safe, then save:
final_panel = final_panel.drop_duplicates(subset=["ric", "Year"])
```

```python
final_panel.to_csv("processed_data/final_data.csv")
```
